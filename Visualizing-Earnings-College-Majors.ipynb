It must be noted that for low values of sample size, the median earnings varies largely.When the sample size is more significant, the median earnings seems to be converging to the correct median salary for the entire population

The same above can be inferred for unemployment_rate

Full_time vs Median seems to agree with Sample_size vs Median

There is no relation between ShareWomen and unemployment_rate

Men make more money than women,as it is evident from the chart.However, the median salary converges to more or less the same value for larger counts of men and women.

So , for popular majors among both men and women, the men and women have same median salary. For majors with more women share, the median salary is low.For all other less popular ones, share of women is less , median salary is more and hence men seems to outdo women.
