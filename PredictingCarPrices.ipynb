```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

def knn_train_test(cols,df,k):
    for col in cols:
        if type(col)=='str':
            print("The passed column is in string format. Regression requires columns in numeric formats only.")
            return

    length = len(df)
    split=int(0.7*length)
    
    df['price']=df['price'].astype(int)
    train = df[0:split]
    
    test=df[split:]
    
    knn=KNeighborsRegressor(n_neighbors=k)
    knn.fit(train[cols],train['price'])
    predictions=knn.predict(test[cols])
    
    rmse=(mean_squared_error(test['price'],predictions))**0.5
    return rmse

```

```python
rmses=[]
for col in num_cols:
    rmses.append(knn_train_test(col,cars,5))

cars_rmse=pd.DataFrame(list(zip(num_cols,rmses)),columns=['Columns','Root Mean Squared Error'])  
cars_rmse.sort_values("Root Mean Squared Error")
```

For default value of k , curb_weight columns has the least root mean squared error and hence performed well among all other columns for 70:30 train test split and KNN Regression.

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

params_k=[1,3,5,7,9]


for k in params_k:
    rmses=[]
    for col in num_cols:
        rmses.append(knn_train_test(col,cars,k))
    cars_rmse=pd.DataFrame(list(zip(num_cols,rmses)),columns=['Columns','Root Mean Squared Error'])  
    print(cars_rmse.sort_values("Root Mean Squared Error"))
    plt.hist(rmses)
    plt.ylim(0,5)
    plt.show()
        
```

from the plots and calculations , k=9 is the best k value among others. 

```python
best_two=['curb_weight','highway_mpg']
best_three=['curb_weight','highway_mpg','engine_size']
best_four=['curb_weight','highway_mpg','engine_size','city_mpg']
best_five=['curb_weight','highway_mpg','engine_size','city_mpg','width']

k=9

rmses=[]

features = [best_two,best_three,best_four,best_five]

for f in features:
    rmses.append(knn_train_test(f,cars,k))
    
best_rmse=pd.DataFrame(list(zip(features,rmses)),columns=['Columns','Root Mean Squared Error'])  
best_rmse.sort_values("Root Mean Squared Error")    
```

best_four list has the lowest root mean squared error and hence contains the best features for KNN and split ratio of 0.7

```python
hyper_params_k=[k for k in range(1,26)]

top_three_best_f =[best_four,best_five,best_three]

group_rmse=[]
for f in top_three_best_f:
    rmses=[]
    for k in hyper_params_k:
        rmses.append(knn_train_test(f,cars,k))
    
    print(f)
    top_rmse=pd.DataFrame(list(zip(hyper_params_k,rmses)),columns=['K','Root Mean Squared Error'])  
    print(top_rmse.sort_values("Root Mean Squared Error"))
    group_rmse.append(list(zip(hyper_params_k,rmses)))
```
