```python
import csv
```

```python
with open("guns.csv","r") as f:
    data=list(csv.reader(f))
```

```python
data[:5]
```

```python
headers=data[0]
data=data[1:]
print(headers)
print(data[:5])
```

```python
year=[row[1] for row in data]
year
```

```python
#Dictionary of death counts by year
year_counts={}

for y in year:
    if y in year_counts:
        year_counts[y]+=1
    else:
        year_counts[y]=1

print(year_counts)        
```

```python
import datetime
```

```python
dates=[datetime.datetime(year=int(row[1]),month=int(row[2]),day=1) for row in data]
```

```python
dates[:5]
```

```python
date_counts={}

for d in dates:
    if d in date_counts:
        date_counts[d]+=1
    else:
        date_counts[d]=1

print(date_counts)
```

```python
sex_counts={}
race_counts={}

for row in data:
    sx = row[5]
    race=row[7]
    
    if sx in sex_counts:
        sex_counts[sx]+=1
    else:
        sex_counts[sx]=1
        
    if race in race_counts:
        race_counts[race]+=1
    else:
        race_counts[race]=1
        
print(sex_counts)
print(race_counts)
```
